## Apply
> python의 for문과 같은 역할

```python
def is_vip(amount):
    #####
    ### pandas로 구현할 수 없는 것들을
    ### 여기서 파이썬 코드를 구현할 수 있다!!
    #####
    return amount

order["amount"].apply(is_vip)
```

```python
def is_vip(amount):
    if amount >= 500:
        return "VIP"
    else:
        return "Normal"
    return amount

order["amount"].apply(is_vip)
```

### 여러개의 Coulmn을 쓰려면

```python
def is_vip(row):
    amount = row["amount"]
    result = row["result"]
    if amount >= 500 and result == "confirmed":
        return "VIP"
    else:
        return "Normal"
order.apply(is_vip, axis="columns")
```
