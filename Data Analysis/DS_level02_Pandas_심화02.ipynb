## 여러 개의 데이터를 하나로 합치기
### 보통 여러개 CSV를 가져와서 데이터를 분석하기 때문에
- e-commerce 데이터 3대장: 유저 데이터, 상품 데이터, 결제 데이터

```python
import pandas as pd
```

```python
url1 = "https://bit.ly/dsa-02-order-201701"
```

```python
order201701 = pd.read_csv(url1, index_col = "id")
order201701
```

```python
url2 = "https://bit.ly/dsa-02-order-201702"
order201702 = pd.read_csv(url2, index_col = "id") #axis ="index" 가 숨어있는 것
order201702
```

- 떨어져서 데이터가 제공되는 경우가 많다.
- 기본적으로 서비스가 크고 트래픽도 많다 -> 하나의 컴퓨터로 커버 불가 -> 분산처리를 해줘야 한다.
- 따라서 데이터를 제공해줄 때에도 작은걸 여러개 제공을 해주는 것이 다운받는게 쉽다. -> 다운받다 실패하는 경우 ;;;;;
- 예를 들어 구글 애널리틱스 같이 분석 툴도 쪼개줌

이 때, concat으로 합쳐준다!

```python
order2017 = pd.concat([order201701, order201702])
order2017.shape
```

```python
url3= "https://bit.ly/dsa-02-order-how"
how = pd.read_csv(url3, index_col = "id")
how
```

```python
pd.concat([order2017,how],axis = "columns") # 왼쪽과 오른쪽 합침
```

```python
# index 를 딱히 지정하지 않았다 -> 순서대로 다합침, 빈 값은 NaN
# 가능한 인덱스를 지정해줘야 꼬이지 않는다.
```

## 다른 데이터 합치기

```python
url = "https://bit.ly/dsa-02-user"
user = pd.read_csv(url, index_col = "id")
user
```

```python
urlp = "https://bit.ly/dsa-02-product"
product = pd.read_csv(urlp, index_col = "id")
product
```

```python
# 일단 공토되는 컬럼을 찾아라
order2017
```

```python
pd.merge(order2017, user, left_on = "user_id", right_on = "id") # 어떤 키로 연결됐는지는 데이터 제공자가 알려줄것임 걱정 ㄴ
```

```python
user["user_id"] = user.index
user
```

```python
# 만약에 왼쪽, 오른쪽 키가 같다?
pd.merge(order2017, user, on = "user_id")
```

겹치는 이름이 없고 오직 공통된 이름이 user_id

### Key를 지정해주지 않으면 암묵적으로 pandas가 자동으로 key를 찾음
### 근데 이름만 우연히 똑같을 뿐 다른 컬럼일수도 주의!

```python
pd.merge(order2017,user)
```

```python
pd.merge(order2017, product, left_on = "product_id", right_on = "id")
```

```python
# price 로 머지를 해버리는 경우
# key는 명시적으로 지정해주자.
pd.merge(order2017, product)
```

```python
pd.merge(order2017, user, on = "user_id")
```

- 양쪽 동시에 존재하는 데이터가 없을 수도
- ex) order에서 userid 8, daegu -> user 에 없음
- user에서 moon 6 -> order에 결제 안해서 없음

```python
pd.merge(order2017, user, on="user_id", how = "inner") #왼쪽 오른쪽 둘다 있는 교집합!
```

```python
pd.merge(order2017, user, on="user_id", how = "outer") # 없는 것은 NaN으로 넣어 합집합!
```

```python
pd.merge(order2017, user, on="user_id", how = "left") # 왼쪽은 살리지만 오른쪽은 없애 vice versa
```
