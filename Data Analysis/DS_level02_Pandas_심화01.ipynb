## Pivot Table
> 대용량 데이터 처리에 속도가 빠르다

```python
# confirmed 일 때 평균 가격 / canceled 일 때 평균 가격
pd.pivot_table(order, index = "state", values = "price")
```

```python
pd.pivot_table(order, index = "address", values = "price")
```

> 중요!!! 
**NaN** 은 이 통계에서 배제된다.

```python
pd.pivot_table(order, index = "state", values = "price", aggfunc = "sum")
```

> 다수 기능을 넣을 수 있는 aggregation function

```python
import numpy as np
pd.pivot_table(order, index = "state", values = "price", aggfunc = [np.mean, np.sum])
```

```python
pd.pivot_table(order, index = ["date","state"], values = "price")
```

## Group by
> Pivot Table 과 기능 유사, 상호 호환 가능
- 하지만 언제언제 각 기능을 쓰는지 알아보자

```python
order["price"].mean()
```

```python
# confirmed 일 때랑 canceled 일 때랑 따로 구하고싶다
# 특정상황이 아닌 여러가지 조건마다 따로 계산해주고 싶다
order.groupby("state")["price"].mean()
```

```python
order.groupby("state")["price"].sum()
```

```python
order.groupby(["date","state"])["price"].sum()
```

- 통계 연산 후 눈으로 확인하고 싶으면 `pivot_table` : 예쁘게 출력되기 때문

- 연산 후 저장 or 추가 연산을 하고싶다면 `groupby`

```python
%timeit pd.pivot_table(order, index = "state", values = "price")
```

```python
%timeit order.groupby("state")["price"].mean()
```

> 출력할 필요가 없다. 무조건 groupby를 쓰자

```python
table = pd.pivot_table(order, index = ["date","state"], values = "price")
table
```

인자에 - 같은 연산기호가 들어가면 안됨

굳이 함수에 컬럼명 - 인자명을 같은걸 써줄 필욘 없음

```python
def is_vip(current-price):
    return current-price

order["current-price"].apply(is_vip)
```

```python
def is_vip(amount):
    return amount

order["price"].apply(is_vip)
```

```python
def is_vip(row):
    if row["state"] == "confirmed" and row["price"] >= 500:
        return "VIP"
    else:
        return "None-VIP"
        

order.apply(is_vip, axis = "columns")
```
